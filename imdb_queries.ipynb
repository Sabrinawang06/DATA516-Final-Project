Look at the number of rows for all tables.

```sql
-- 2 join
SELECT titleType, primaryTitle, startYear, genres, averageRating, numVotes
FROM title_basics tb, title_ratings tr
WHERE tb.tconst = tr.tconst
	AND genres LIKE '%Comedy%'
    AND genres LIKE '%Romance%'
    AND numVotes > 10000
ORDER BY averageRating DESC
```

```sql
-- 2 join agg
SELECT startYear, AVG(averageRating) as avgOfaverageRating
FROM title_basics tb, title_ratings tr
WHERE tb.tconst = tr.tconst
	AND genres LIKE '%Comedy%'
    AND genres LIKE '%Romance%' 
    AND numVotes > 10000
GROUP BY startYear
ORDER BY avgOfaverageRating DESC
```

```sql
-- 4 join
SELECT primaryName, titleType, primaryTitle, genres, averageRating
FROM title_basics tb, title_ratings tr, title_principals tp, name_basics nb
WHERE tb.tconst = tr.tconst
    AND tb.tconst = tp.tconst
    AND tp.nconst = nb.nconst
    AND startYear >= 1994
    AND startYear <= 1994 + 10
    AND primaryName IN ('Jennifer Aniston',
                        'Courteney Cox', 
                        'Matthew Perry',
                        'Lisa Kudrow', 
                        'David Schwimmer', 
                        'Matt LeBlanc')
```

```sql
-- 4 join agg
SELECT primaryName, titleType, count(*) as numTitles, AVG(averageRating) as avgOfaverageRating
FROM title_basics tb, title_ratings tr, title_principals tp, name_basics nb
WHERE tb.tconst = tr.tconst
    AND tb.tconst = tp.tconst
    AND tp.nconst = nb.nconst
    AND startYear >= 1994
    AND startYear <= 1994 + 10
    AND primaryName IN ('Jennifer Aniston',
                        'Courteney Cox', 
                        'Matthew Perry',
                        'Lisa Kudrow', 
                        'David Schwimmer', 
                        'Matt LeBlanc')
GROUP BY primaryName, titleType
ORDER BY numTitles DESC;
```

```sql
-- 6 join
SELECT primaryName, tb2.titleType, tb2.primaryTitle, tp2.category, tb2.genres, tr.averageRating
FROM name_basics nb, title_principals tp1, title_basics tb1, title_principals tp2, title_basics tb2, title_ratings tr
WHERE nb.nconst = tp1.nconst
	AND tb1.tconst = tp1.tconst
    AND tb1.primaryTitle = 'Friends'
    AND tb1.titleType = 'tvSeries'
    AND tb1.startYear = 1994
    AND tp1.category IN ('actor', 'actress')
    AND nb.nconst = tp2.nconst
    AND tp2.tconst = tb2.tconst
    AND tb2.tconst = tr.tconst
```

```sql
-- 6 join agg
SELECT nb.primaryName, tp2.category, count(*) as numTitles, AVG(tr.averageRating) as avgOfaverageRating
FROM name_basics nb, title_principals tp1, title_basics tb1, title_principals tp2, title_basics tb2, title_ratings tr
WHERE nb.nconst = tp1.nconst
	AND tb1.tconst = tp1.tconst
    AND tb1.primaryTitle = 'Friends'
    AND tb1.titleType = 'tvSeries'
    AND tb1.startYear = 1994
    AND tp1.category IN ('actor', 'actress')
    AND nb.nconst = tp2.nconst
    AND tp2.tconst = tb2.tconst
    AND tb2.tconst = tr.tconst
GROUP BY nb.primaryName, tp2.category
ORDER BY avgOfaverageRating DESC;
```

```sql
-- nested 1 level
SELECT
	(CASE WHEN numEpisodes >= 200 then 'A. 200+'
    	 WHEN numEpisodes >= 100 then 'B. 100-199'
         WHEN numEpisodes >= 50 then 'C. 50-99'
         WHEN numEpisodes >= 25 then 'D. 25-49'
         WHEN numEpisodes >= 2 then 'E. 2-24'
         WHEN numEpisodes = 1 then 'F. 1'
	END) episodeRange,
    COUNT(*) AS totalInRange
FROM(SELECT tp.nconst, COUNT(*) AS numEpisodes
	 FROM title_episode te, title_basics tb, title_principals tp
     WHERE te.parentTconst = tb.tconst
     	AND te.tconst = tp.tconst
     	AND tb.primaryTitle = 'Friends'
     	AND tb.titleType = 'tvSeries'
     	AND tb.startYear = 1994
	GROUP BY tp.nconst) ne
GROUP BY (CASE WHEN numEpisodes >= 200 then 'A. 200+'
    	 WHEN numEpisodes >= 100 then 'B. 100-199'
         WHEN numEpisodes >= 50 then 'C. 50-99'
         WHEN numEpisodes >= 25 then 'D. 25-49'
         WHEN numEpisodes >= 2 then 'E. 2-24'
         WHEN numEpisodes = 1 then 'F. 1'
	END)
ORDER BY episodeRange;
```

```sql
-- nested 2 level
SELECT
	(CASE WHEN numEpisodes >= 200 then 'A. 200+'
    	 WHEN numEpisodes >= 100 then 'B. 100-199'
         WHEN numEpisodes >= 50 then 'C. 50-99'
         WHEN numEpisodes >= 25 then 'D. 25-49'
         WHEN numEpisodes >= 2 then 'E. 2-24'
         WHEN numEpisodes = 1 then 'F. 1'
	END) AS episodeRange,
    COUNT(*) AS totalInRange
FROM(SELECT tp.nconst, COUNT(*) AS numEpisodes
	 FROM title_episode te, title_basics tb, title_principals tp,
     	  (SELECT nconst, COUNT(*) AS numCredits
			FROM title_principals
			GROUP BY nconst
			HAVING COUNT(*) > 10) nc
     WHERE te.parentTconst = tb.tconst
     	AND te.tconst = tp.tconst
     	AND nc.nconst = tp.nconst
     	AND tb.primaryTitle = 'Friends'
     	AND tb.titleType = 'tvSeries'
     	AND tb.startYear = 1994
	GROUP BY tp.nconst) ne
GROUP BY (CASE WHEN numEpisodes >= 200 then 'A. 200+'
    	 WHEN numEpisodes >= 100 then 'B. 100-199'
         WHEN numEpisodes >= 50 then 'C. 50-99'
         WHEN numEpisodes >= 25 then 'D. 25-49'
         WHEN numEpisodes >= 2 then 'E. 2-24'
         WHEN numEpisodes = 1 then 'F. 1'
	END)
ORDER BY episodeRange;
```
